****AI for Green Plates – Agri-Food CO₂ Emission Classification API****
**Project Overview**

This project develops and deploys a machine learning model that classifies global agri-food systems into low, medium, or high CO₂ emission intensity based on country-level agricultural and supply chain data.

The model also provides explainable insights into which agricultural practices—such as fertilizer use, rice cultivation, manure management, energy consumption, and transport—contribute most to emissions.

The API can be integrated into dashboards, sustainability monitoring tools, or decision-making workflows to support policymakers, environmental organizations, and agri-food companies in reducing carbon footprints across food systems.

 **Problem Description**

Accurately identifying high-impact sources of agricultural emissions is critical for designing sustainable food policies, incentivizing low-carbon farming practices, and achieving climate goals.

This project provides a machine learning–powered API that classifies emission intensity from historical agricultural and supply chain data, enabling:
 + Evidence-based policy and intervention planning.
 + Identification of the most impactful practices contributing to emissions.
 + Monitoring and reporting of progress toward greener food systems.

The service is designed for integration into dashboards, monitoring tools, or automated sustainability workflows, making insights actionable and easy to use.

 **Dataset Information**

The dataset contains country-level data capturing agricultural processes and food production features that influence CO₂ emissions. It can be used to classify emission intensity, analyze drivers of emissions, and guide sustainability efforts.

Key columns include:

+ country: Name of the country.
+ fertilizer_use: Amount of fertilizer applied (kg per hectare).
+ rice_cultivation: Rice production volume or area cultivated.
+ manure_management: Practices related to manure handling.
+ crop_residues: Crop residue management practices.
+ energy_consumption: Energy used in agricultural production (e.g., electricity, fuel).
+ food_transport: Emissions related to transporting food products.
+ CO2_emission: Total CO₂ emissions (used to create the target label).
+ emission_level: Target label for classification (Low, Medium, High).

Dataset Source:
Kaggle – Agri-Food CO₂ Emission Dataset
(https://www.kaggle.com/datasets/alessandrolobello/agri-food-co2-emission-dataset-forecasting-ml)

```python
# Handling missing values
# Separate categorical & numerical
cat_cols = ["Area"]
num_cols = [col for col in df.columns if col not in cat_cols]

# Fill missing
df[cat_cols] = df[cat_cols].fillna("NA")
df[num_cols] = df[num_cols].fillna(0.0)
```
